```python
from sage.modules.fp_graded.steenrod.module import SteenrodFPModule as fpmod
from fphomspaces import *
```

```python
def draw_chart(chart, size=12):
    r'''
    
    Utility function which draws the ext charts below.
      
    INPUT:
    - ``chart`` -- a list of lists of non-negative integers.
      Each inner list represents a row in the chart.  Each
      integer in the inner list will be represented by that
      number of black dots within the same cell in the chart.
    '''
    lst = []
    nx = 0
    ny = len(chart)

    # Create a flat list of coordinates for every dot
    # to appear in the final chart.
    for y, l in enumerate(chart):
        nx = max(nx, len(l))
        for x, p in enumerate(l):
            for i in range(p):
                lst.append((x+0.5+i*0.125,y+0.5))

    # Make visible points from the coordinate list.
    pts = points(lst, color="black", size=24)
    
    # Create a background grid.
    grid = Graphics()
    for i in range(0,nx+1,1):
        grid += line([(i,0),(i,ny)], color="#666")
        if i < nx:
            grid += text(i, (i+0.5,0-0.3), color="#666")
    for i in range(0,ny+1,1):
        grid += line([(0,i),(nx,i)], color="#666")
        if i < ny:
            grid += text(i, (-0.3, i+0.5), color="#666")
            
    # Show everything.
    show(grid + pts, aspect_ratio=1.0, figsize=size, axes=False)
```

```python
# ko
# Ext_A(A tensor_A1 F2, F2) ~ Ext_A1(F2, F2)f

A1 = SteenrodAlgebra(2, profile=(2,1))

F2_A1 = SteenrodFPModule(A1, [0], [[Sq(1)], [Sq(2)]])

res = F2_A1.resolution(30, verbose=False)
```

```python
chart = [ [ext(res, F2_A1, s, t+s) for t in range(0, 60)] for s in range(0,len(res)-1) ]
draw_chart(chart)
```

```python
# ku
# Ext_A(A tensor_E1 F2, F2) ~ Ext_E1(F2, F2)

E1 = SteenrodAlgebra(2, profile=(1,1))

F2_E1 = SteenrodFPModule(E1, [0], [[Sq(1)], [Sq(0,1)]])

res = F2_E1.resolution(30, verbose=False)
```

```python
chart = [ [ext(res, F2_E1, s, t+s) for t in range(0, 60)] for s in range(0,len(res)-1) ]
draw_chart(chart)
```
